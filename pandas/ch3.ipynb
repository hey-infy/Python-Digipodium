```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('Canada.xlsx', skiprows = 20, skipfooter = 2, sheet_name='Canada by Citizenship' )    #skiprows to remove unwanted rows , skipfooter to remove unwanted columns 
df
```

```python
df.value_counts()
```

```python
df.info()
```

```python
df.describe(include='object')   # include object is used to describe the text columns only excluding number values 
```

```python
#Data Cleaning             (is 80% Data analysis)
cols_to_drop = ['Type', 'Coverage', 'AREA', 'REG', 'DEV']   #Making a list of all columns to be dropped
df.drop(columns=cols_to_drop, inplace = True)     #inplace is done to save the cleaned data in memory (but first run a test to verify the cleaning process)
```

```python
print(df.columns.to_list())
```

```python
df.rename( columns={
    'OdName':'Country', 
    'AreaName':'Continent',
    'RegName':'Region',
    'DevName':'Status'}, inplace=True
)
```

```python
df.head(3)
```

```python
years = list(range(1980,2014))
print(years)
```

```python
#Adding new column
df['Total'] = df[years].sum(axis=1)   #Summing all the values horizontally i.e. axis = 1 and df['total'] is new column to be added 
```

```python
df.head(3)
```

```python
df.set_index('Country', inplace=True)    #here country name is set as index number (index is placed on x axis)
```

```python
df.sort_values(by=1980, ascending=False).head(25)[1980].plot(kind='bar', figsize=(20,5))    #Sorting the data according to 1980 in desc order , fetch top 25 from 1980 and the plot a bar graph
```

```python
df.sort_values(by=2000, ascending=False).head(5)[2000].plot(kind='bar', figsize=(20,5))
```

```python
df.sort_values('Total', ascending=False, inplace=True)  #Sort according to total immigration in all the years 
df.head()
```

```python
df.loc['Japan', years].plot(kind='line', style='bo-', title='Japan Immigration Trend', figsize=(15,5))      #Analyse the data according to years of any country and then plotting a line chart against it
```

```python
df.loc['India', years].plot(kind='bar', color='red', title='India Immigration Trend', alpha=0.5, figsize=(15,3))
```

For making multiple entries in one graph plot each individual entry one after the other and title in last one is read, also function like legend is read only in last line (Legend is used along with label to add individual title in mixed graph)

```python
df.loc['Japan', years].plot(kind='line',style='bo-',label = 'Japan', figsize=(15,5))
df.loc['France', years].plot(kind='line', style='ro-',label = 'France', figsize=(15,5))
df.loc['Afghanistan', years].plot(kind='line', style='go-',label = 'Afghanistan', title='Immigration Trend', figsize=(15,5)).legend()
```
