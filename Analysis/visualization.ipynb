```python
#random line plot (profit,loss,trends)
x=np.linspace(0,10,100)
y=np.random.randn(100)    #randn :-  random number like randint is random integer
print(x,y)
```

```python
plt.figure(figsize=(15,6), dpi=300)   #figsize used to resize graph and dpi for clear and sharp lines(dots per inch)
_ = plt.plot(x,y, linewidth=3, color='red', marker='o')  #_ is used for removing the memory location line, linewidth for line thick, color for line color 
```

```python
#random bar plot
plt.figure(figsize=(15,4), dpi=200)
x = np.random.randint(0,10,25)
y = np.random.randint(0,10,25)
plt.grid(False)        #false means gridlines not show, else dont write this line  
_ = plt.bar(x,y,color='pink')
```
